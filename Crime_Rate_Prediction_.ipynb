# Verify The all Information gather are correct or not.

````python
dataset = pd.DataFrame()
dataset['State'] = load_dataset['State']
dataset['Year'] = load_dataset['Year']
dataset['Total_Cases'] = load_dataset['Victims_Total']
````

````python
# @title State vs Year

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(dataset['State'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(dataset, x='Year', y='State', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
````

# It can used to transform non-numerical labels to numerical labels.

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
````

````python
dataset.iloc[:,0] = le.fit_transform(dataset.iloc[:,0])
````

````python
dataset.sample(10)
````

````python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
````

````python
x.head()
````

````python
y.head()
````

# Split arrays or matrices into random train and test subsets.

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
````

# Create a linear regression model

````python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
````

# Fit the model to the dataset

````python
regressor.fit(x_train,y_train)
````

````python
y_pred = regressor.predict(x_test)
````
